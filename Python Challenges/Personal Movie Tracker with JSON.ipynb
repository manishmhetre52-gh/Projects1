```python
class IMDB:
    def __init__(self):
        self.movie_list = movie_list
    def add_movie(self):
        title = input("Enter Movie Title: ").strip().title()
        for i in self.movie_list:
            if i["title"] == title:
                return f"{title} movie already exists in the storage"
        genre = input("Enter Movie Genre: ").strip().title()
        rating = float(input("Enter Movie Rating: "))
        self.movie_list.append({"title": title, "genre": genre, "rating": rating})
        with open("movies.json","w") as w:
            json.dump(self.movie_list,w,indent=4)
        return f"{title} movie added to the storage"
    def view_movies(self):
        for i in self.movie_list:
            print(f"Name:{i['title']} - Genre:{i['genre']} - Rating:{i['rating']}")
    def search_movie(self):
        choice = input("Search by Title or Genre?: ").strip().lower()
        if choice == "title":
            title = input("Enter Movie Title: ").strip().title()
            for i in self.movie_list:
                if i["title"] == title:
                    print(f"Name:{i['title']} - Genre:{i['genre']} - Rating:{i['rating']}")
            return f"No match found for movie with the title: {title}"
        elif choice == "genre":
            genre = input("Enter Movie Genre: ").strip().title()
            for i in self.movie_list:
                if i["genre"] == genre:
                    print(f"Name:{i['title']} - Genre:{i['genre']} - Rating:{i['rating']}")
            return f"No match found for movie with the genre: {genre}"
    def main(self):
        while True:
            menu = input("Welcome to the Movie Search Menu!\nChoose Your Option.\n1.Add Movie.\n2.View All Movies.\n3.Search Movie.\n4.Exit\n").strip()
            if menu == "1":
                print(self.add_movie())
            elif menu == "2":
                self.view_movies()
            elif menu == "3":
                self.search_movie()
            elif menu == "4":
                print("Thanks for using this program")
                break
            else:
                return "Invalid Option"
```

```python
imdb = IMDB()
```

```python
imdb.main()
```
