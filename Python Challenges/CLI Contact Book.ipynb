```python
class PhoneBook:
    def __init__(self):
        self.contacts = "contacts.csv"
    def add_contact(self):
        name = input("Enter name: ").strip().title()
        with open(self.contacts,'r') as csvfile:
            reader = csv.DictReader(csvfile)
            for row in reader:
                if row['name'] == name:
                    return f"Contact with {name} already exists."
        phone = input("Enter phone: ").strip()
        if not len(phone) == 10 or not phone.isdigit():
            return f"Phone number must be 10 digits."
        email = input("Enter email: ").strip()
        with open(self.contacts,'a') as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow([name,phone,email])
            return f"Contact with {name} has been added."
    def view_contacts(self):
        with open(self.contacts,'r') as csvfile:
            reader = csv.DictReader(csvfile)
            rows = list(reader)
            print("\n{:<20} {:<15} {:<30}".format("Name", "Phone", "Email"))
            print("-" * 65)
            for row in rows:
                print("{:<20} {:<15} {:<30}".format(
                    row["name"],
                    row["phone"],
                    row["email"]
                ))
            print()
    def search_contact(self):
        name = input("Enter name: ").strip().title()
        with open(self.contacts,'r') as csvfile:
            reader = csv.DictReader(csvfile)
            rows = list(reader)
            print("\n{:<20} {:<15} {:<30}".format("Name", "Phone", "Email"))
            print("-" * 65)
            for row in rows:
                if row['name'] == name:
                    return "\n{:<20} {:<15} {:<30}".format(row["name"],row["phone"],row["email"])
            return f"Contact with {name} does not exist."
    def main(self):
        while True:
            menu = input("Welcome to the phonebook!\nChoose Your Option\n1.Add Contact\n2.View Contacts\n3.Search Contact\n4.Exit")
            if menu == "1":
                print(self.add_contact())
            elif menu == "2":
                self.view_contacts()
            elif menu == "3":
                print(self.search_contact())
            elif menu == "4":
                print("Thank you for using this program")
                break
            else:
                print("Please enter a valid option")
```
